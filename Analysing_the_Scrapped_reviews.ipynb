```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & pictographs
        "\U0001F680-\U0001F6FF"  # Transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # Flags
        "\U00002500-\U00002BEF"  # Chinese characters
        "\U00002702-\U000027B0"
        "\U00002702-\U000027B0"
        "\U000024C2-\U0001F251"
        "\U0001f926-\U0001f937"
        "\U00010000-\U0010ffff"
        "\u2640-\u2642"
        "\u2600-\u2B55"
        "\u200d"
        "\u23cf"
        "\u23e9"
        "\u231a"
        "\ufe0f"  # Dingbats
        "\u3030"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
# for topic modeling we have
# 1. Base imports
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation, NMF
```

```python
# Count Vectorizer
count_vect = CountVectorizer(stop_words='english')
dtm = count_vect.fit_transform(df['review'])

# LDA Model
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(dtm)

# Topic Display Function
def display_topics(model, feature_names, n_top_words=5):
    for topic_idx, topic in enumerate(model.components_):
        print(f"🔹 LDA Topic {topic_idx+1}: " + ", ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))

display_topics(lda, count_vect.get_feature_names_out())
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F
```

```python
# Load the tokenizer and model
model_name = "bhadresh-savani/bert-base-go-emotion"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
def predict_emotions(text, top_k=3):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", truncation=True)

    # Get model predictions
    with torch.no_grad():
        logits = model(**inputs).logits
        probs = torch.sigmoid(logits)[0]  # Apply sigmoid for multi-label classification

    # Get top k emotions
    top_probs, top_indices = torch.topk(probs, k=top_k)
    labels = [model.config.id2label[idx.item()] for idx in top_indices]
    scores = [prob.item() for prob in top_probs]

    # Return results as a list of tuples (label, score)
    return list(zip(labels, scores))
```

```python
# text = "I love the quality of this razor; it's perfect for my skin."
emotions = predict_emotions(review)

for label, score in emotions:
    print(f"{label}: {score:.4f}")
```

```python
def emotion(text):
  abc = predict_emotions(text)  # Use the function parameter 'text'
  return abc

df["emotions_review"] = df["review"].apply(emotion)
```

```python
from transformers import pipeline

# Load a general emotion classifier that includes empathy-related labels
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", top_k=None)
```

```python
# Sample input for empathy detection
text = "I'm really sorry to hear you're going through such a tough time. That must be incredibly hard."

# Run prediction
results = classifier(text)
for r in results[0]:
    print(f"{r['label']}: {r['score']:.4f}")
```
